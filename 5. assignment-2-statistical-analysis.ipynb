```python
import sqlite3
import pandas as pd
```

```python
conn1 = sqlite3.connect("/kaggle/input/aggregate-framework-data-into-sqlite-database/crux_data.db")
conn2 = sqlite3.connect("/kaggle/input/web-scraper-2-preliminary-data-processing/store.db")
```

```python
MAX_RANK = 10000000
pd.options.display.max_columns = 300
```

```python
pd.read_sql(f"SELECT * FROM CrUXData WHERE rank <= {MAX_RANK} LIMIT 5", conn1)
```

```python
pd.read_sql(f"SELECT * FROM websiteFrameworkData WHERE rank <= {MAX_RANK} LIMIT 5", conn2)
```

```python
df1 = pd.read_sql(f"SELECT * FROM CrUXData WHERE rank <= {MAX_RANK}", conn1)
df2 = pd.read_sql(f"SELECT * FROM websiteFrameworkData WHERE rank <= {MAX_RANK}", conn2)

df = pd.merge(df1, df2, on = "domain", how = "inner")
df.head(5)
```

```python
df1 = pd.read_sql(f"SELECT * FROM CrUXData WHERE rank <= {MAX_RANK}", conn1)
del df1
del df2
df["rank"] = df["rank_x"]
del df["idx"] # redundant
del df["rank_x"] # redundant
del df["rank_y"] # redundant
del df["yyyymmdd"] # redundant
del df["yyyymm"] # redundant

assert not (df['desktopDensity'] * df['phoneDensity'] * df['tabletDensity']).any()
df['deviceDensity'] = df['desktopDensity'] + df['phoneDensity'] + df['tabletDensity']
del df['desktopDensity'] #in deviceDensity now
del df['phoneDensity'] #in deviceDensity now
del df['tabletDensity'] #in deviceDensity now

df.head(5)
```

```python
df["whatrunsDateUnixMillis"] = pd.to_datetime(df["whatrunsDateUnixMillis"].astype("float64"), unit='ms')
df["date"] = pd.to_datetime(df["date"])
df["protocol"] = df["protocol"].astype("category")
df["device"] = df["device"].astype("category")
df["siteIsOnline"] = df["siteIsOnline"] == 1
df.head(5)
```

```python
df.to_parquet('website_data.parquet', compression='snappy')
```
